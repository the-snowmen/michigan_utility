```python
import os
from PIL import Image



def downscale_images(input_folder, output_folder, size):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path)
            img_resized = img.resize(size, Image.LANCZOS)
            output_path = os.path.join(output_folder, filename)
            img_resized.save(output_path)
            print(f"{filename} has been resized and saved to {output_folder}")

input_folder = r"C:\Users\MHuang\OneDrive - Everstream\Desktop\MikeDHuang_2024\Machine Learning\basedata\Tester\Test_Hit480"
output_folder = r"C:\Users\MHuang\OneDrive - Everstream\Desktop\MikeDHuang_2024\Machine Learning\basedata\Tester\Test_Hit1080"

downscale_images(input_folder, output_folder, size=(1080,1080))
```
